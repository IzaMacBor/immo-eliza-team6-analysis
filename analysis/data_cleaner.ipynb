**EDITING TEXT, CHANGING DATA TYPE, ADDING UNITS**

```python
import pandas as pd

data = r'C:\Users\izama\Desktop\immo_scraper_merged.csv'
data_scraped = pd.read_csv(data, sep = ',')

#Replace 'True' and 'False' value on binary values
columns_for_change_01 = ['Equipped_kitchen', 'Furnished', 'Open_fire', 'Swimming_Pool'] #Columns for change
data_scraped[columns_for_change_01] = data_scraped[columns_for_change_01].replace({False:0, True:1})

#Add m² for area columns 
columns_for_change_m2 = ['Living_area', 'Terrace', 'Garden'] #Columns for change

for col in columns_for_change_m2:
    data_scraped[col] = pd.to_numeric(data_scraped[col], errors='coerce').fillna(0).astype('Int64') #Change string value to an integer 

for col in columns_for_change_m2:
    data_scraped[col] = data_scraped[col].apply(lambda x: f"{x} m²" if pd.notna(x) and x != 0 else x) #Add m² to the area values

#Replace "0" with "Null"
columns_for_change_null = ['Terrace', 'Garden']
data_scraped[columns_for_change_null] = data_scraped[columns_for_change_null].replace({0: 'null'})

#Edit text in the locality column
data_scraped['locality_name'] = data_scraped['locality_name'].str.title()

#Edit text in the type of sale column
data_scraped['Type_of_sale'] = data_scraped['Type_of_sale'].str.replace('_', ' ').str.capitalize()

#Edit the number of facades
data_scraped['Number_of_facades'] = pd.to_numeric(data_scraped['Number_of_facades'], errors='coerce').fillna(0).astype('Int64')

#Edit the price column
data_scraped['Price'] = data_scraped['Price'].fillna(0).astype(int)

#Edit text in the state of building column
data_scraped['State_of_building'] = data_scraped['State_of_building'].str.replace('_', ' ').str.capitalize()

#Edit text in SUB(?)type of property column
data_scraped['Type_of_property'] = data_scraped['Type_of_property'].str.replace('_', ' ').str.capitalize()

data_scraped[500:520]
```
